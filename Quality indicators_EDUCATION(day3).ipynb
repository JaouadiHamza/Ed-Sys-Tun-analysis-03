```python
from matplotlib.ticker import PercentFormatter
plt.figure(figsize=(20,10))
data10.sort_values(by='Value',ascending=True)
data10['cumperc'] = data10['Value'].cumsum()/data10['Value'].sum()*100
fig, ax = plt.subplots()
ax.bar(data10['LOCATION'], data10['Value'], color='steelblue')
ax2 = ax.twinx()
ax2.plot(data10['LOCATION'], data10['cumperc'], color='red', marker="o", ms=2)
ax2.yaxis.set_major_formatter(PercentFormatter())
```
